```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
df = pd.read_csv('/Users/purva/Desktop/Data Science Boot Camp/examples/avocado 2.csv')
df = df[(df.region != 'TotalUS')&(df.region != 'Midsouth')&(df.region != 'Northeast')&(df.region != 'SouthCentral')&(df.region != 'West')&(df.region != 'Southeast')&(df.region != 'California')]
con = df.loc[lambda f2: df['type'] == 'conventional', :]
org = df.loc[lambda f2: df['type'] == 'organic', :]
plt.subplot(1, 2, 1)
plt.plot(org.groupby('year')['AveragePrice'].mean(), color = 'green')
plt.plot(con.groupby('year')['AveragePrice'].mean(), color = 'red')
plt.xlabel('Year')
plt.ylabel('Average price of Avocadoes')
plt.tight_layout()

plt.subplot(1, 2, 2)
plt.plot(org.groupby('year')['Total Volume'].mean(), color = 'green')
plt.plot(con.groupby('year')['Total Volume'].mean(), color = 'red')
plt.xlabel('Year')
plt.ylabel('Volume produced')
plt.tight_layout()
plt.show()
```

First graph above shows the trend in average avocado prices for last 4 years for conventional as well as organic avocadoes. It can be seen that, the prices suddenly reach peak in year 2017 for organic as well as conventional avocadoes. However, the prices of avocadoes drop suddenly in 2018.
Second graph shows the trend in volume produced in last 4 years. For organic avocadoes, the volueme produced is constant throughout 4 years. In case of conventional avocadoes, there is slight increase in volume from 2015 to 2016 and constant in 2017 and then suddenly goes up in 2018.
However, it is difficult to relate the price of avocadoes and the volueme produced.

```python
plt.figure(figsize=(20, 5))

plt.subplot(1, 2, 1)
plt.plot(org.groupby('year')['4046'].mean(), color = 'green')     #Small Hass
plt.plot(org.groupby('year')['4225'].mean(), color = 'red')      #Large Hass
plt.plot(org.groupby('year')['4770'].mean(), color = 'blue')    #Extra Large Hass
plt.xlabel('Year')
plt.ylabel('Prices for PLU')

plt.subplot(1, 2, 2)
plt.plot(con.groupby('year')['4046'].mean(), color = 'green')     #Small Hass
plt.plot(con.groupby('year')['4225'].mean(), color = 'red')      #Large Hass
plt.plot(con.groupby('year')['4770'].mean(), color = 'blue')    #Extra Large Hass
plt.xlabel('Year')
plt.ylabel('Prices for PLU')
plt.show()
```

The first above graph shows price variation for three three types of organic avocadoes. It can be seen that price for Extra Large Hass is almost constant. For Large hass it initial goes down and remains constant. For Small Hass, there is no particular trend seen.
In second graph above, variation for three types of conventional avocadoes can be seen.Again it can be seen that price for Extra Large Hass is almost constant.For small hass and large hass, there is no particular trend but the prices go down initially and then start going up.

```python
df1 = df.groupby('year').aggregate(np.sum)
i = np.arange(4)

plt.bar(i,df1['Small Bags'], 0.35)
plt.bar(i,df1['Large Bags'], 0.35)
plt.bar(i,df1['XLarge Bags'], 0.35)
plt.xticks(i, (2015,2016,2017,2018))
plt.xlabel('Year')
plt.ylabel('Number of bags')
plt.show()
## print(fig_size)
```

The above graph shows the total number of bags per sold in each year. 2017 is the year with highest number of bags. 2018 is the year with least number of bags. Incomplete data for 2018 might be the reason for least number of bags.
